```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.io
import math
import sklearn
import sklearn.datasets

import opt_utils
import testCase
import updateFunction

%matplotlib inline 

plt.rcParams['figure.figsize'] = (7.0, 4.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'


train_X, train_Y = opt_utils.load_dataset(is_plot = False)

def model(X, Y, layer_dims, optimizer, mini_batch_size = 64, learning_rate = 0.0007, beta = 0.9, beta1 = 0.9, beta2 = 0.99, epsilon = 1e-8, num_epochs = 10000, print_cost = True, is_plot = True):
    """
    功能:
    模拟利用三种不同的优化算法的三层神经网络传输
    参数:
    X:输入数据
    Y:输入数据的标签
    layer_dims:网络每层的节点分布
    optimizer:优化器类型，["gd"|"momentum"|"adam"]
    learning_rate:学习率
    mini_batch_size:mini_batch的大小
    beta:动量优化的超参数
    beta1:用于计算梯度后的指数衰减的估计的超参数
    beta2:用于计算平方梯度后的指数衰减的估计的超参数
    epsilon:Adam避免分母为0的超参数
    num_epochs:整个训练集的迭代次数
    print_cost:是否打印
    is_plot:是否画图
    返回值:
    parameters:更新后的参数字典
    """
    L = len(layer_dims)
    costs = []
    t = 0
    #adam算法的迭代次数
    seed = 10
    
    parameters = opt_utils.initialize_parameters(layer_dims)
    
    
    if optimizer == "gd":
        pass
    elif optimizer == "momentum":
        v = updateFunction.initialize_velocity(parameters)
    elif optimizer == "adam":
        v, s = updateFunction.initialize_adam(parameters)
    else:
        print("optimizer参数错误，程序退出。")
        exit(1)
    
    
    #开始学习
    for i in range(num_epochs):
        #每次不同的seed随机生成minibatches，每次数据的分配都不一样
        seed = seed + 1 
        minibatches = updateFunction.random_mini_batches(X, Y, mini_batch_size, seed)
        
        for minibatch in minibatches:
            (mini_batch_X, mini_batch_Y) = minibatch
            
            #前向传播
            A3, cache = opt_utils.forward_propagation(mini_batch_X, parameters)
            
            #计算误差
            cost = opt_utils.compute_cost(A3, mini_batch_Y)
            
            #反向传播
            grads = opt_utils.backward_propagation(mini_batch_X, mini_batch_Y, cache)
            
            #更新参数
            if optimizer == "gd":
                parameters = updateFunction.update_parameters_with_gd(parameters, grads, learning_rate)
            elif optimizer == "momentum":
                parameters, v = updateFunction.update_parameters_with_momentun(parameters, grads, v, beta, learning_rate)
            elif optimizer == "adam":
                t = t + 1
                parameters, v, s = updateFunction.update_parameters_with_adam(parameters, grads, v, s, t, learning_rate, beta1, beta2, epsilon)
        
        if i % 100 == 0:
            costs.append(cost)
            if print_cost and i % 1000 == 0:
                print("第" + str(i) + "次遍历整个数据集，当前误差值：" + str(cost))
    
    if is_plot:
        plt.plot(costs)
        plt.ylabel('cost')
        plt.xlabel('epochs (per 100)')
        plt.title("Learning rate = " + str(learning_rate))
        plt.show()
    
    return parameters

layer_dims = [train_X.shape[0],5,2,1]

#parameters = model(train_X, train_Y, layer_dims, optimizer="gd")

#predictions = opt_utils.predict(train_X, train_Y, parameters)

#plt.title("Model with Gradient Descent optimization")
#axes = plt.gca()
#axes.set_xlim([-1.5, 2.5])
#axes.set_ylim([-1, 1.5])
#opt_utils.plot_decision_boundary(lambda x: opt_utils.predict_dec(parameters, x.T), train_X, train_Y)

#parameters = model(train_X, train_Y, layer_dims, beta = 0.9, optimizer = "momentum", is_plot = True)

#predictions = opt_utils.predict(train_X, train_Y, parameters)

#plt.title("Model with Momentum optimization")
#axes = plt.gca()
#axes.set_xlim([-1.5, 2.5])
#axes.set_ylim([-1, 1.5])
#opt_utils.plot_decision_boundary(lambda x: opt_utils.predict_dec(parameters, x.T), train_X, train_Y)

parameters = model(train_X, train_Y, layer_dims, optimizer="adam",is_plot=True)
preditions = opt_utils.predict(train_X,train_Y,parameters)


plt.title("Model with Adam optimization")
axes = plt.gca()
axes.set_xlim([-1.5, 2.5])
axes.set_ylim([-1, 1.5])
opt_utils.plot_decision_boundary(lambda x: opt_utils.predict_dec(parameters, x.T), train_X, train_Y)
```
