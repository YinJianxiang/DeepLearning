```python
import numpy as np
from matplotlib import pyplot as plt
from tf_utils import load_dataset
import torch
from torch import nn
from torch.utils.data import DataLoader, TensorDataset
import imageio
import cv2

np.random.seed(1)
torch.manual_seed(1)

#线性函数
def linear_function():
    X = torch.from_numpy(np.random.randn(3,1))
    W = torch.from_numpy(np.random.randn(4,3))
    b = torch.from_numpy(np.random.randn(4,1))
    Y = torch.add(torch.matmul(W, X), b)
    
    return Y.numpy()

print("----------test linrear_function----------")
print("linear_fonction = ",linear_function())

def sigmoid(z):
    z = torch.tensor(z, dtype=torch.float32)
    sig = torch.sigmoid(z)
    
    return sig

def cost(logits, labels):
    """
    参数:
    logits:没有activate的预测
    labels:实际值
    """
    input = sigmoid(logits.numpy())
    
    loss = nn.BCELoss(reduction = 'none')
    cost = loss(input, labels)
    
    return cost

print("-----test loss-----")
logits = sigmoid(np.array([0.2, 0.4, 0.7, 0.9]))
labels = torch.Tensor([0, 0, 1, 1])
cost = cost(logits, labels)
print('cost = ', cost.numpy())

#独热代码
def one_hot_martix(label, C):
    """
    参数：
    label:标签向量
    C:分类数
    返回值：
    one_hot:独热矩阵
    """
    m = label.shape[0]
    one_hot = torch.zeros(size = (C,m))
    
    for l in range(m):
        one_hot[label[l], l] = 1 
        
    return one_hot

print("-----test one_hot_matrix-----")
labels = np.array([1,2,3,0,2,1])
one_hot = one_hot_martix(labels, C=4)
print('one_hot = ' + str(one_hot))    

def ones(shape):
    one = torch.ones(shape)
    return one.numpy()
    
```

```python
X_train_orig, Y_train_orig, X_test_orig, Y_test_orig, classes = load_dataset()

"""
index = 0
plt.imshow(X_train_orig[index])
plt.show()
print("y = ", np.squeeze(Y_train_orig[:, index]))
"""

#把图像数据展开
X_train_flatten = X_train_orig.reshape(X_train_orig.shape[0], -1)
X_test_flatten = X_test_orig.reshape(X_test_orig.shape[0], -1)

#归一化
X_train = X_train_flatten / 255
X_test = X_test_flatten / 255

Y_train = Y_train_orig.T
Y_test = Y_test_orig.T

print(Y_train_orig.shape)
def neural_net():
    """
    网络框架
    linear->relu->linear->relu->linear->softmax
    W1:[25, 12288]
    b1:[25, 1]
    W2:[12, 25]
    b2:[12, 1]
    W3:[6, 12]
    b3:[6, 1]
    """
    net = nn.Sequential(
        nn.Linear(12288, 25),
        nn.ReLU(),
        nn.Linear(25, 12),
        nn.ReLU(),
        nn.Linear(12, 6),
        nn.LogSoftmax(dim = 1)
        #dim = 0每一列所有元素和为1
        #dim = 1每一行所有元素和为1
    )
    return net

net = neural_net()
#print(net)

def initialize_params(net):
    """
    初始化参数
    """
    for i in range(len(net)):
        layer = net[i]
        if isinstance(layer, nn.Linear):
            #w用Xavier初始化
            nn.init.xavier_uniform_(layer.weight, gain = torch.nn.init.calculate_gain('relu'))
            #gain根据激活函数类型设定
            nn.init.constant_(layer.bias, 0)
            #用value填充Tensor
            #print('W' + str(i//2+1) + ' = ' + str(layer.weight.dtype) + ' ' + str(layer.weight.shape))
            #print('b' + str(i//2+1) + ' = ' + str(layer.bias.dtype) + ' ' + str(layer.bias.shape))

    return net

#Net = initialize_params(net)

def data_loader(X_train, Y_train, batch_size = 32):
    train_id = TensorDataset(torch.from_numpy(X_train).float(), torch.squeeze(torch.from_numpy(Y_train)))
    #数据对应标签
    #print("train ", train_id)
    train_loader = DataLoader(train_id, batch_size=batch_size, shuffle=True)
    
    return train_loader


def model(X_train, Y_train, X_test, Y_test, lr = 0.001, epochs = 1500,batch_size = 32, print_cost = True, is_plot = True):
    #载入数据
    train_loader = data_loader(X_train, Y_train, batch_size)
    #创建网络 
    net = neural_net()
    #成本函数
    cost_func = nn.NLLLoss()
    #优化器Adam
    optimizer = torch.optim.Adam(net.parameters(), lr=lr,betas=(0.9, 0.999))
    
    costs = []
    
    m = X_train.shape[0]
    num_batch = m / batch_size
    
    for epoch in range(epochs):
        epoch_cost = 0
        
        for step, (batch_x, batch_y) in enumerate(train_loader):
            #前向传播
            output = net(batch_x)
            #计算成本
            cost = cost_func(output, np.squeeze(batch_y))
            epoch_cost += cost.data.numpy() / num_batch
            #梯度归0
            optimizer.zero_grad()
            #反向传播
            cost.backward()
            #更新参数
            optimizer.step()
        if print_cost and epoch % 5 == 0:
            costs.append(epoch_cost)
            if epoch % 100 == 0:
                print("epoch = " + str(epoch) + "      epoch_cost = " + str(epoch_cost))
    
    if is_plot:
        plt.plot(np.squeeze(costs))
        plt.ylabel('cost')
        plt.xlabel('iterations (per tens)')
        plt.title("Learning rate =" + str(lr))
        plt.show()
        
    # 保存学习后的参数
    torch.save(net.state_dict(), 'net_params.pkl')
    print('参数已保存到本地pkl文件。')

    #计算训练集预测的结果
    net.load_state_dict(torch.load('net_params.pkl'))
    output_train = net(torch.from_numpy(X_train).float())
    pred_Y_train = torch.max(output_train, dim=1)[1].data.numpy()
    # 计算测试集预测的结果
    output_test = net(torch.from_numpy(X_test).float())
    pred_Y_test = torch.max(output_test, dim=1)[1].data.numpy()
    # 训练集准确率
    print('Train Accuracy: %.2f %%' % float(np.sum(np.squeeze(Y_train) == pred_Y_train)/m*100))
    # 测试集准确率
    print('Test Accuracy: %.2f %%' % float(np.sum(np.squeeze(Y_test) == pred_Y_test)/X_test.shape[0]*100))
    
    
    return net

Net = model(X_train, Y_train, X_test, Y_test, lr=0.0001, epochs=1500, batch_size=32, print_cost=True, is_plot=True)       
```

```python
def test_local_picture(img_path, y, train_net, num_px = 64):
    image = np.array(imageio.imread(img_path))
       
    image_cut = cv2.resize(image, dsize=(num_px, num_px))[:, :, :3].reshape(1, -1)
    
    x_test = image_cut/255
    
    y_test = np.array(y).reshape(1,1)
    output = train_net(torch.from_numpy(x_test).float())
    y_pred = torch.max(output, dim=1)[1].data.numpy()
    y_pred = int(np.squeeze(y_pred))
    
    print("y = %i, predicted %i %s" % (y, y_pred, ('Correct.' if y == y_pred else 'Wrong.')))
    

train_net = neural_net()
train_net.load_state_dict(torch.load('net_params.pkl'))

IMG_PATH = 'test_images/%i.png'
for i in range(6):
    img_path = IMG_PATH % i
    #print(img_path)
    test_local_picture(img_path, i, train_net)
    
    
```
