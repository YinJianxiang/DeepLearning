```python
%matplotlib inline
import numpy as np
import h5py
import matplotlib.pyplot as plt
import testCases 
from dnn_utils import sigmoid, sigmoid_backward, relu, relu_backward 
import lr_utils

np.random.seed(1) #指定随机种子

#初始化参数
def initialize_parameters(n_x, n_h, n_y):
    """
    功能：
    实现两层神经网络参数的初始化
    参数：
    n_x：输入层节点数
    n_h：隐藏层节点数
    n_y：输出层节点数
    
    返回值：
    parameters：包含参数的字典
        W1：权重矩阵，维度(n_h,n_x)
        b1：偏向量，维度(n_h,1)
        W2：权重矩阵，维度(n_y,n_h)
        b2：偏向量，维度(n_y,1)
    """
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros((n_h, 1))
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros((n_y, 1))
    
    assert(W1.shape == (n_h, n_x))
    assert(b1.shape == (n_h, 1))
    assert(W2.shape == (n_y, n_h))
    assert(b2.shape == (n_y, 1))
        
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters

def initialize_parameters_deep(layers_dims):
    """
    实现多层神经网络参数的初始化
    参数：
    layers_dims：包含网络每层的节点数目的列表
    返回值：
    parameters：包含Wi，bi参数的字典
        Wi：权重矩阵，维度(layers_dims [i], layers_dims [i-1])
        bi：偏向量，维度(layers_dims [i], 1)
    """
    np.random.seed(3)#指定随机种子
    
    parameters = {}
    L = len(layers_dims)
    
    for l in range(1, L):
        parameters["W" + str(l)] = np.random.randn(layers_dims[l], layers_dims[l - 1]) * 0.01
        parameters["b" + str(l)] = np.zeros((layers_dims[l], 1))
        
        assert(parameters["W" + str(l)].shape == (layers_dims[l], layers_dims[l - 1]))
        assert(parameters["b" + str(l)].shape == (layers_dims[l], 1))

    return parameters

#前向传播
#linear
#linear->activate,Relu or Sigmopig
#[linear->Relu](L-1) ->[linear->Sigmoid]

#linear
#Zl = Wl * A(l-1) + bl
#其中A0 = X

def linear_forward(A, W, b):
    """
    功能：实现前向的线性传播
    参数：
    A：来自上一层的激活函数输出，维度(n[l-1], m)
    W：权重矩阵，维度(n[l], n[l - 1])
    b：偏向量，维度(n[l], 1)
    返回值：
    Z：激活函数的输入
    cache：包含A,W,b的字典，用于反向传播
    """
    Z = np.dot(W, A) + b
    assert(Z.shape == (W.shape[0], A.shape[1]))
    cache = (A, W, b)
    
    return Z, cache


#linear->activate
#A[l] = g(Z[l])
#sigmoig(Z) = 1/(1+e^{-(WA+b)})
#Relu = max(0, Z)

def linear_activation_forward(A_prev, W, b, activation):
    """
    功能：实现线性输出后的激活
    参数：
    A_prev：经过线性传播后的输出，维度(n[l], m)
    W：权重矩阵，维度(n[l], n[l-1])
    b：偏向量，维度(n[l],1)
    activation：激活函数，字符串类型
    返回值：
    A：激活函数输出
    cache：包含linear_cache和activation_cache的字典，用于反向传播
    """
    if activation == "sigmoid":
        Z, linear_cache = linear_forward(A_prev, W, b)
        A, activation_cache = sigmoid(Z)
    elif activation == "relu":
        Z, linear_cache = linear_forward(A_prev, W, b)
        A, activation_cache = relu(Z)
    
    assert(A.shape == (W.shape[0], A_prev.shape[1]))
    cache = (linear_cache, activation_cache)
    
    return A,cache

#L层前向传播
def L_model_forward(X, parameters):
    """
    功能：
    实现L层前向传播（L-1层Relu激活，1层sigmoid）
    参数：
    X：输入数据，维度(n_x, m)
    parameters：经过初始化Wl,bl的神经网络参数
    返回值：
    AL：输出值
    caches： linear_relu_forward（）的每个cache（有L-1个，索引为从0到L-2）
             linear_sigmoid_forward（）的cache（只有一个，索引为L-1）
    """
    caches = []
    A = X
    
    L = len(parameters) // 2
    
    for l in range(1, L):
        A_prev = A
        A, cache = linear_activation_forward(A_prev, parameters['W' + str(l)], parameters['b' + str(l)], "relu")
        caches.append(cache)
    
    AL, cache = linear_activation_forward(A, parameters['W' + str(L)], parameters['b' + str(L)], "sigmoid")
    caches.append(cache)

    assert(AL.shape == (1,X.shape[1]))
    
    return AL,caches

def compute_cost(AL, Y):
    """
    功能：计算成本函数
    参数：
    AL：对应，维度(1，m)
    Y：数据标签的标签，维度(1,m)
    返回值：
    cost：交叉熵成本
    """
    m = Y.shape[1]
    cost = -np.sum(np.multiply(np.log(AL),Y) + np.multiply(np.log(1 - AL), 1 - Y)) / m
    
    cost = np.squeeze(cost)
    #让[[12]]变成12 
    assert(isinstance(cost, float))

    return cost
                       
#反向传播
#linear_backward
#linear_backward activation(Relu or Sigmoid)
#[linear->Relu](L-1) -> [linear->Sigmoid]
                       
#dW[l] = (1/m) dZ[l] A[l-1]T
#db[l] = (1/m) * np.sum(dZ[l], axis = 1, keepdims = True)                       
#dA[l-1] = W[l]T dZ[l]

def linear_backward(dZ, cache):
    """
    功能：
    实现反向传播的线性部分
    参数：
    dZ：Z[l]的梯度值
    cache：前向传播时缓存用于反向传播的参数(A_prev, W，b)
    返回值：
    dA_prev：A[l-1]的梯度值
    dW：W[l]的梯度值
    db：b[l]的梯度值
    """
    A_prev, W, b = cache
    m = A_prev.shape[1]
    dW = np.dot(dZ, A_prev.T) / m
    db = np.sum(dZ, axis = 1, keepdims = True) / m
    dA_prev = np.dot(W.T, dZ)                   
                       
    assert (dA_prev.shape == A_prev.shape)
    assert (dW.shape == W.shape)
    assert (db.shape == b.shape)
    
    return dA_prev, dW, db


#linear->activation
#dZ[l] = dA[l] * g'(Z[l])
#分别用sigmoid_backward or relu_backward

def linear_activation_backward(dA, cache, activation):
    """
    功能：
    实现反向传播的激活函数
    参数：
    dA：A[l]的梯度值
    cache：存储的用于有效计算反向传播的值
    activation：激活函数，字符串类型
    返回值：
    dA_prev：A[l-1]的梯度值
    dW：W[l]的梯度值
    db：b[l]的梯度值
    """
    linear_cache, activation_cache = cache
    if activation == "relu":
        dZ = relu_backward(dA, activation_cache)
        dA_prev, dW, db = linear_backward(dZ, linear_cache)
    elif activation == "sigmoid":
        dZ = sigmoid_backward(dA, activation_cache)
        dA_prev, dW, db = linear_backward(dZ, linear_cache)
    
    return dA_prev, dW, db

#L层反向传播
#[linear_backward->sigmoid](1)->[linear_backward->relu](L-1)
#dA[L] = sum(-(y(i)/a(i)) + (1-y(i))/(1-a(i)))
def L_model_backward(AL, Y, caches):
    """
    功能：
    实现L层反向传播，[linear_backward->sigmoid](1)->[linear_backward->relu](L-1)
    参数：
    Al：正向传播输出向量
    Y：输出数据的标签向量
    caches：正向传播中的存储在反向传播需要用到的参数
    返回值：
    grads：包含梯度的字典，dA[l],dW[l],db[l]
    
    """
    grads = {}
    L = len(caches)
    #网络层数
    m = AL.shape[1]
    Y = Y.reshape(AL.shape)
    
    dAL = -(np.divide(Y, AL) - np.divide(1 - Y, 1 - AL))
    
    current_cache = caches[L - 1]
    grads["dA" + str(L-1)], grads["dW" + str(L)], grads["db" + str(L)] = linear_activation_backward(dAL, current_cache, "sigmoid")
    
    for l in reversed(range(L-1)):
        current_cache = caches[l]
        dA_prev_temp, dW_temp, db_temp = linear_activation_backward(grads["dA" + str(l+1)], current_cache, "relu")
        grads["dA" + str(l)] = dA_prev_temp
        grads["dW" + str(l + 1)] = dW_temp
        grads["db" + str(l + 1)] = db_temp
    
    return grads
        
#更新参数
#/alpha 为学习率
#W[l] = W[l] - /alpha dW[l]
#b[l] = b[l] - /alpha db[l]

def update_parameters(parameters, grads, learning_rate = 0.01):
    """
    功能：
    利用梯度下降更新参数
    参数：
    parameters：包含神经网络参数的字典
    grads：包含梯度值的字典
    learning_rate：学习率
    返回值：
    parameters：更新参数后的字典
    """
    L = len(parameters) // 2 #整除
    for l in range(L):
        parameters["W" + str(l + 1)] = parameters["W" + str(l + 1)] - learning_rate * grads["dW" + str(l + 1)]
        parameters["b" + str(l + 1)] = parameters["b" + str(l + 1)] - learning_rate * grads["db" + str(l + 1)]
        
    return parameters

def two_layer_model(X, Y, layers_dims, learning_rate = 0.075, num_iterations = 3000, print_cost = False, isPlot = True):
    """
    功能：实现两层神经网络
    参数：
    X：输入数据，维度(n_x, m)
    Y：输入数据标签，维度(1, m)
    layer_dims：包含层数的向量,(n_x, n_y, n_h)
    learning_rate：学习率
    num_iterations:迭代次数
    print_cost：是否打印成本值
    isPlot：是否画出绘制图谱
    返回值：
    parameters：包含W1,b1,W2,b2参数的字典
    """
    np.random.seed(1)#指定随机种子
    grads = {}
    costs = []
    
    (n_x, n_h, n_y) = layers_dims
    
    #初始化参数
    parameters = initialize_parameters(n_x, n_h, n_y)
    
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    
    for i in range(0, num_iterations):
        #前向传播
        A1, cache1 = linear_activation_forward(X, W1, b1, "relu")
        A2, cache2 = linear_activation_forward(A1, W2, b2, "sigmoid")
        
        # 计算成本
        cost = compute_cost(A2, Y)
        
        #后向传播
        dA2 = -(np.divide(Y, A2) - np.divide(1 - Y, 1 - A2))
        
        #向后传播
        dA1, dW2, db2 = linear_activation_backward(dA2, cache2, "sigmoid")
        dA0, dW1, db1 = linear_activation_backward(dA1, cache1, "relu")
        
        grads["dW1"] = dW1
        grads["db1"] = db1
        grads["dW2"] = dW2
        grads["db2"] = db2
        
        #保存grads
        parameters = update_parameters(parameters, grads, learning_rate)
        W1 = parameters["W1"]
        b1 = parameters["b1"]
        W2 = parameters["W2"]
        b2 = parameters["b2"]
        
        #每100次记录、打印一次成本值
        if i % 100 == 0:
            costs.append(cost)
            if print_cost:
                print("第", i ,"次迭代，成本值为：" ,np.squeeze(cost))
    
    #画图
    if isPlot:
        plt.plot(np.squeeze(costs))
        plt.ylabel('cost')
        plt.xlabel('iterations (per tens)')
        plt.title("Learning rate =" + str(learning_rate))
        plt.show()
    
    return parameters

def predict(X, Y, parameters):
    """
    功能：
    通过神经网络对测试集数据进行测试
    参数：
    X：测试集数据
    Y：测试集标签
    parameters：包含神经网络参数的字典
    返回值：
    Y_prediction:测试集预测
    """
                  
    m = X.shape[1]
    layer = len(parameters) // 2
    Y_prediction = np.zeros((1, m))
    
    #进行前向传播
    prediction, caches = L_model_forward(X, parameters)              
    
    #对预测结果进行判断
    for i in range(0, prediction.shape[1]):
        if prediction[0, i] > 0.5:
            Y_prediction[0, i] = 1
        else:
            Y_prediction[0, i] = 0

    print("准确度为: "  + str(float(np.sum((Y == Y_prediction))/m)))
                  
    return Y_prediction
"""               
#读入数据
print("---------------两层神经网络测试---------------")
train_set_x_orig , train_set_y , test_set_x_orig , test_set_y , classes = lr_utils.load_dataset()

train_x_flatten = train_set_x_orig.reshape(train_set_x_orig.shape[0], -1).T 
test_x_flatten = test_set_x_orig.reshape(test_set_x_orig.shape[0], -1).T

train_x = train_x_flatten / 255
train_y = train_set_y
test_x = test_x_flatten / 255
test_y = test_set_y

n_x = train_x.shape[0]
n_h = 7
n_y = 1
layers_dims = (n_x, n_h, n_y)

parameters = two_layer_model(train_x, train_set_y, layers_dims, num_iterations = 2500, print_cost=True,isPlot=True)

predict_train = predict(train_x, train_y, parameters)
predict_test = predict(test_x, test_y, parameters)
"""

#L层神经网络传播

def L_layer_model(X, Y, layers_dims, learning_rate = 0.075, num_iterations = 3000, print_cost = False, isPlot = False):
    """
    功能：
    实现L层神经网络，[linear->relu](L-1) -> [linear->sigmoid](1)
    参数：
    X:输入数据，维度(n_x, m)
    Y:输入数据的标签，维度(1, m)
    layer_dim:层数的向量,(n_y,n_h,...,n_,n_y)
    learning_rate：学习率
    num_iterations:迭代次数
    print_cost：是否打印成本值
    isPlot：是否画出绘制图谱
    返回值：
    parameters：包含神经网络参数的字典
    """
    np.random.seed(1)
    costs = []
    
    parameters = initialize_parameters_deep(layers_dims)
    
    for i in range(0, num_iterations):
        AL, caches = L_model_forward(X, parameters)
        
        cost = compute_cost(AL,Y)
    
        grads = L_model_backward(AL, Y, caches)
        
        parameters = update_parameters(parameters, grads, learning_rate)
        
        #每100次记录、打印一次成本值
        if i % 100 == 0:
            costs.append(cost)
            if print_cost:
                print("第", i ,"次迭代，成本值为：" ,np.squeeze(cost))
    
    #画图
    if isPlot:
        plt.plot(np.squeeze(costs))
        plt.ylabel('cost')
        plt.xlabel('iterations (per tens)')
        plt.title("Learning rate =" + str(learning_rate))
        plt.show()
    
    return parameters


print("---------------L层神经网络测试---------------")
train_set_x_orig , train_set_y , test_set_x_orig , test_set_y , classes = lr_utils.load_dataset()

train_x_flatten = train_set_x_orig.reshape(train_set_x_orig.shape[0], -1).T 
test_x_flatten = test_set_x_orig.reshape(test_set_x_orig.shape[0], -1).T

train_x = train_x_flatten / 255
train_y = train_set_y
test_x = test_x_flatten / 255
test_y = test_set_y
layers_dims = [12288, 20, 7, 5, 1] #  5-layer model


parameters = L_layer_model(train_x, train_y, layers_dims, num_iterations = 3000, print_cost = True, isPlot = True)

predict_train = predict(train_x, train_y, parameters)
predict_test = predict(test_x, test_y, parameters)


def print_mislabeled_images(classes, X, Y, Y_prediction):
    """
    绘制预测和实际不同的图像。
    X:输入数据，维度(n_x, m)
    Y:输入数据的标签，维度(1, m)
    Y_prediction:根据神经网络得到的预测
    """
    a = Y + Y_prediction
    mislabeled_indices = np.asarray(np.where(a == 1))
    plt.rcParams['figure.figsize'] = (40.0, 40.0) # set default size of plots
    num_images = len(mislabeled_indices[0])
    for i in range(num_images):
        index = mislabeled_indices[1][i] 
        
        plt.subplot(2, num_images, i + 1)
        plt.imshow(X[:,index].reshape(64,64,3), interpolation='nearest')
        plt.axis('off')
        plt.title("Prediction: " + classes[int(Y_prediction[0,index])].decode("utf-8") + " \n Class: " + classes[Y[0,index]].decode("utf-8"))

print("---------------预测错误图片---------------")
print_mislabeled_images(classes, test_x, test_y, predict_test)
```
